```python
import pandas as pd

df = pd.read_csv('BlinkIT Grocery Data.csv')
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# Imputing missing values in Item Weight column with the median rating of the item type

df['Item Weight'] = df.groupby('Item Type')['Item Weight'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({
    'LF': 'Low Fat',
    'low fat': 'Low Fat',
    'reg': 'Regular'
})
```

```python
df['Item Fat Content'].unique()
```

```python
total_sales_million = round(df['Total Sales'].sum() / 1_000_000, 2)
print(total_sales_million)
```

```python
df['Total Sales'] = df['Total Sales'].round(2)
```

```python
df['Item Visibility'] = df['Item Visibility'].round(3)
```

```python
df.head()
```

```python
labels = ['Low Visibility', 'High Visibility']

df['Visibility Category'] = pd.qcut(
    df['Item Visibility'],
    q=2,
    labels=labels
)
```

```python
df.to_csv("BlinkIT_Cleaned.csv", index=False)
```
